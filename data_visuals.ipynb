```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/Users/Janzen/Documents/test_scores.csv")
```

```python
df.info()
```

```python
df.drop(['student_id', 'classroom'], axis=1, inplace=True)
```

```python
df.describe()
```

```python
#use histogram to view distribution of data
plt.figure(figsize=(5,6))
plt.hist(df['posttest'])
plt.title('posttest')
```

```python
#use histogram to view distribution of data
plt.figure(figsize=(5,6))
plt.hist(df['pretest'])
plt.title('pretest')
```

```python
pd.crosstab(df["school_setting"], df["lunch"]).plot(kind="bar",figsize=(10,7))
plt.show()
```

```python
pd.crosstab(df["school_type"], df["lunch"]).plot(kind="bar",figsize=(10,7))
plt.show()
```

```python
pd.crosstab(df["teaching_method"], df["school_setting"]).plot(kind="bar",figsize=(10,7), title="teaching_method and school_Setting")
plt.show()
```

```python
pd.crosstab(df["teaching_method"], df["school_type"]).plot(kind="bar",figsize=(10,7), title="teaching_method and school_type")
plt.show()
```

```python
#get mean by each lunch
df.groupby(["lunch"])["posttest"].mean().to_frame().T.plot(kind="bar", figsize=(10,4), title="lunch and posttest mean")

#remove any x axis label
plt.axes().axes.get_xaxis().set_visible(False)
plt.show()
```

```python
#get posttest mean by each school_setting
df.groupby(["school_setting"])["posttest"].mean().to_frame().T.plot(kind="bar", figsize=(10,4), title="school_setting posttest mean")
plt.axes().axes.get_xaxis().set_visible(False)
plt.show()
```

```python
#get posttest mean by each teaching_method
df.groupby(["teaching_method"])["posttest"].mean().to_frame().T.plot(kind="bar", figsize=(10,4), title="teaching_method and posttest mean")
plt.axes().get_xaxis().set_visible(False)
plt.show()
```

```python
#get posttest mean by each school_type
df.groupby(["school_type"])["posttest"].mean().to_frame().T.plot(kind="bar", figsize=(10,4), title="school_type and posttest mean")
plt.axes().get_xaxis().set_visible(False)
plt.show()
```
